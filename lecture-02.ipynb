```r
x = seq(from = 0, to = 1*pi, length.out = 100)
```

```r
y = 10 * sin(x) + rnorm(100, 0)
```

```r
plot(y)
```

```r
df <- data.frame(x, y)
```

```r
plot(x, y, col = 'red')
fit1 <- loess('y ~ x', df, span = 0.10)
lines(x, predict(fit1), col = 'blue')
fit2 <- loess('y ~ x', df, span = 0.25)
lines(x, predict(fit2), col = 'green')
fit3 <- loess('y ~ x', df, span = 0.50)
lines(x, predict(fit3), col = 'purple')
```
